# 2025.3.21 选课社区爬虫

    爬虫内容：课程名称+课程评分+课程学分+评分人数

## 爬取数据（截止到2025.3.21共590页，11792门课程）

```python
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# 设置Edge浏览器的选项
edge_options = Options()
# 伪装user-agent
edge_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
# 无头模式（可选，如果不想看到浏览器窗口可以启用）
# edge_options.add_argument('--headless')

# 创建Edge浏览器驱动实例
driver = webdriver.Edge(options=edge_options)
url = 'https://course.sjtu.plus/latest'
driver.get(url)
time.sleep(30)
# 总页数
total_pages = 590

# 用于存储课程信息的列表
course_data = []

for page in range(1, total_pages + 1):
    url = f"https://course.sjtu.plus/courses?page={page}&size=20"
    driver.get(url)
    time.sleep(1) # 实测发现，需要用time.sleep()来确保稳定性，否则会漏读
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "ul > li"))
    )  # 等待页面加载

    # 遍历每一页的课程
    for i in range(1, 21):
        try:
            # 尝试获取课程名字
            course_name_xpath1 = f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div[1]/div[1]/a'
            course_name_xpath2 = f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div/div[1]/a'
            try:
                course_name_element = driver.find_element(By.XPATH, course_name_xpath1)
            except:
                course_name_element = driver.find_element(By.XPATH, course_name_xpath2)
            course_name = course_name_element.text
            

            # 尝试获取课程评分
            course_score_xpath1 = f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div[2]/div[1]/span/strong'
            course_score_xpath2 = f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/span'
            try:
                course_score_element = driver.find_element(By.XPATH, course_score_xpath1)
            except:
                course_score_element = driver.find_element(By.XPATH, course_score_xpath2)
            course_score = course_score_element.text
            

            # 尝试获取课程学分信息
            course_credit_xpaths = [
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div[1]/div[2]/div/div[2]/span',
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div/div[2]/div/div[2]/span',
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div[1]/div[2]/div/div/span',
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div/div[2]/div/div/span'
            ]
            course_credit = None
            for xpath in course_credit_xpaths:
                try:
                    course_credit_element = driver.find_element(By.XPATH, xpath)
                    course_credit = course_credit_element.text
                    break
                except:
                    continue
            

            # 尝试获取课程评分人数信息
            course_rating_count_xpaths = [
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/div[2]/div[2]/span',
                f'//*[@id="root"]/div/main/div[2]/div[2]/div/div[2]/div/div[1]/div/ul/li[{i}]/span'
            ]
            course_rating_count = None
            for xpath in course_rating_count_xpaths:
                try:
                    course_rating_count_element = driver.find_element(By.XPATH, xpath)
                    course_rating_count = course_rating_count_element.text
                    break
                except:
                    continue
            
            
            # 将课程信息添加到列表中
            course_data.append({
                "课程名字": course_name,
                "课程分数": course_score,
                "课程学分": course_credit,
                "课程评分人数": course_rating_count
            })


        except:
            # 如果某一个课程信息获取失败，说明可能是最后一页不足20个课程的情况，直接跳出循环
            break

    print(f"已爬取第 {page} 页, 总共包含{len(course_data)} 条数据")

# 关闭浏览器
driver.quit()
```

## 写入excel表格

```python
# 定义 Excel 文件路径
excel_file_path = 'course_info.xlsx'

# 写入 Excel 文件
df = pd.DataFrame(course_data)
df.to_excel(excel_file_path, index=False)
```
